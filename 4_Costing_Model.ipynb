```
#Import spark library
from pyspark.sql.functions import *
import pyspark.sql.functions as F
from pyspark.sql.types import *

#ipmort python libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```
df = spark.read.table('rgb_db.act_demand_sales_forecast')
```

```
df.display()
```

```
(df
 .filter( (df.fct_volume.isNull()) | (df.lag == 4) ) #keep rows with missing fct and fct 4 weeks out
 .withColumn('over_forecast', #overproduction
             F.when(df.demand.isNull(), df.fct_volume) #forecast with no demand
             .when(df.fct_volume.isNull(), 0)
             .when(df.demand > df.fct_volume, 0)
             .otherwise(df.fct_volume - df.demand) #forecast > demand
            )
 .withColumn('under_forecast', #underproduction, report as negative
             F.when(df.fct_volume.isNull(), df.demand*-1) #demand with no forecast
             .when(df.demand.isNull(), 0)
             .when(df.fct_volume > df.demand, 0)
             .otherwise(df.fct_volume - df.demand) #demand > forecast
            )
 .display()
)
```
